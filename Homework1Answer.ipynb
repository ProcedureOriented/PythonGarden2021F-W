```python
'''
1. 设计程序，接收两个整数x和y，并立即输出这两个整数的和、差、乘积和商
要求：不考虑意外情况，设计最简单的程序

样例：
请输入x：
6
请输入y（不能为0）：
2
【输出】
x+y=8
x-y=4
x*y=12
x/y=3
'''

x = eval(input('请输入x：'))
y = eval(input('请输入y（不能为0）：'))

print('x+y=', x+y, sep='')
print('x-y=', x-y, sep='')
print('x*y=', x*y, sep='')
print('x/y=', x/y, sep='')

# 也可以不加sep参数
```

```python
'''
5. 设计程序，依次接收三个系数a、b、c，输出方程a*X**2+b*X+c=0的实根。

样例1：
请输入a：
2
请输入b：
1
请输入c：
-1
【输出】
0.5,-1

样例2：
请输入a：
2
请输入b：
1
请输入c：
1
【输出】
没有实数根
'''

a = eval(input('请输入a：'))
b = eval(input('请输入b：'))
c = eval(input('请输入c：'))

dt = b**2 - 4*a*c
if dt>0:
    dt = dt**0.5
    x1 = (-b + dt)/2/a
    x2 = (-b - dt)/2/a
    print(x1, ',', x2, sep='')
elif dt==0:
    # 相同实根也应输出两个结果，更规范一些
    x1=x2=-b/2/a
    print(x1,",",x2)

    # 直接输出一个实根
    # print(-b/a/2)
else:
    print('没有实数根')
```

```python
'''
6. 设计程序，接收一个整数（0~100）作为成绩，输出评级。
评级标准：大于等于90为优，大于等于80不及90为良，大于等于70不及80为中，大于等于60不及70为及格，小于60为不及格。
额外思考：如何使判断条件更简单？

样例：
请输入成绩（0~100）：
90
【输出】
优
'''

score=eval(input("请输入成绩（0~100）："))
if score<60:
    print("不及格")
elif score<70:
    print("及格")
elif score<80:
    print("中")
elif score<90:
    print("良")
else:
print("优")

# 倒着排会多写等号
# if score>=90:
#     print('优')
# elif score>=80:
#     print('良')
# elif score>=70:
#     print('中')
# elif score>=60:
#     print('及格')
# else:
#     print('不及格')
```

```python
# 6. 精简算法
ls1=[60,70,80,90,101]
ls2=['不及格','及格','中','良','优']
score = eval(input('请输入成绩（0~100）：'))
for i in range(len(ls1)):
    if score < ls1[i]:
        print(ls2[i])
        break
```

```python
'''
8. 设计一个判断闰年的程序，接收一个年份，输出是否是闰年
规则：当年份能被4整除而不能被100整除是闰年，但是能被400整除时也是闰年。

样例：
请输入年份：
2012
【输出】
闰年

样例：
请输入年份：
1900
【输出】
平年

样例：
请输入年份：
2000
【输出】
闰年
'''

year=eval(input("请输入年份："))
if year%4!=0:
    print("平年")
else:
    if year%100!=0:
        print("闰年")
    else:
        if year%400==0:
            print("闰年")
        else:
            print("平年")
```

```python
# 8. 精简算法（训练一下自己能否理清逻辑运算？）
year = eval(input('请输入年份：'))
if year%400 == 0 or (year%100 and year%4 == 0):
    print('闰年')
else:
    print('平年') 
```
